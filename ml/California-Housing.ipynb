### More about ocean proximity

```python
raw_data['ocean_proximity'].value_counts()
```

```python
raw_data.loc[:, ['ocean_proximity', 'median_house_value']].groupby('ocean_proximity').mean()
```

```python
ignore = raw_data.loc[:, ['ocean_proximity', 'median_house_value']].boxplot(by='ocean_proximity', figsize = (20, 10))
```

```python
stats.f_oneway(raw_data.loc[raw_data['ocean_proximity'] == '<1H OCEAN', ['median_house_value']], 
               raw_data.loc[raw_data['ocean_proximity'] == 'INLAND', ['median_house_value']],
               raw_data.loc[raw_data['ocean_proximity'] == 'NEAR BAY', ['median_house_value']],
               raw_data.loc[raw_data['ocean_proximity'] == 'NEAR OCEAN', ['median_house_value']],
              )
```

ANOVA suggests that the group means are significantly different from each other.

- Inland houses seem to be significantly cheaper than the rest.
- There are only 5 'Island' houses. They are all quite expensive.
- The other three categories have overlapping prices and therefore they alone cannot determine the price. 
- There are a lot of outliers in 'Inland' category.

```python
ignore = raw_data.loc[:, ['ocean_proximity', 'median_income']].boxplot(by='ocean_proximity', figsize = (20, 10))
```

- All categories have a large number of outliers on the upper side.
- There is no great income disparity among the categories.

```python
stats.f_oneway(raw_data.loc[raw_data['ocean_proximity'] == '<1H OCEAN', ['median_income']], 
               raw_data.loc[raw_data['ocean_proximity'] == 'INLAND', ['median_income']],
               raw_data.loc[raw_data['ocean_proximity'] == 'NEAR BAY', ['median_income']],
               raw_data.loc[raw_data['ocean_proximity'] == 'NEAR OCEAN', ['median_income']],
              )
```

Although the box plot suggests that there is no great income parity, statistical test suggests otherwise.

### Correlation between median income and median house price

```python
ax = raw_data.loc[:, ['median_income', 'median_house_value']].plot.scatter(x='median_income',
                                                                               y='median_house_value',
                                                                               figsize=(20, 10))
ax.set_title('Median house value by median income')
ax.set_xlabel('Median income')
ax.set_ylabel('Median house price')
```

```python
stats.pearsonr(raw_data['median_income'], raw_data['median_house_value'])
```

Pearson's test suggests that there is significant correlation between the two variables. It's also visually evident.
